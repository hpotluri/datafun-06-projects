## Section 9-Map: Use map() to transform some of the data.

#### Converting Severity value from UNK to unknown

```python
print(weather_events_df.Severity.unique())

weather_events_df.Severity = weather_events_df.Severity.map(lambda x : 'Unknown' if x == 'UNK' else x)

weather_events_df.Severity.unique()
```
